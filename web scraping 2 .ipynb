```python
import requests
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup

```

```python
Name= []
Price=[]
Brand=[]
Offer=[]
scrol=0
for scrol in range(50): # for loop to get through all the pages by Assume the maximum number of pages for a category 

    try:
        scrol=scrol+1
        response = requests.get("https://shop.tamimimarkets.com/category/baby-1&layoutType=GRID&sorting=NEW_FIRST&page="+str(scrol))
        soup = BeautifulSoup(response.text, 'html.parser')
                #print(response.status_code)
        x=0
        allProduct = soup.find(class_="ProductCollection__StyledContainer-sc-33fjso-2 jgRiGw")
        product = allProduct.find_all(class_="Product__StyledA-sc-13egllk-0 lnZJxA")
        for x in range (10000):
            try:
                subProduct= product[x] 
                name = [n.get_text() for n in subProduct.find_all(class_='Text-sc-1bsd7ul-0 eKTmJK')]
                price = [p.get_text() for p in subProduct.find_all(class_='Text-sc-1bsd7ul-0 loTFgW')]
                brand = [b.get_text() for b in subProduct.find_all(class_='Text-sc-1bsd7ul-0 fOTkAa')]
                offer = [o.get_text() for o in subProduct.find_all(class_='Price__StyledDel-sc-1svix5t-0 iUgmeI')]
                x=x+1
                Name.append(name)
                Price.append(price)
                Brand.append(brand)
                Offer.append(offer)

                    # If the products end in this category
            except IndexError:
                break
           # If the number of pages ends for this category
    except AttributeError:
        break
        print('end baby category')
            
baby = pd.DataFrame({
    "Category": 'baby',
    "Name": Name,
    "Price": Price,
    "Brand": Brand,
    "Offer": Offer,
})        
baby   
```

```python
categoryName=[]
Category= [
 'breakfast',
 'snacks--confectionery',
 'oil--ghee',
 'rice--pasta',
 'water--beverages',
 'baking',
 'frozen',
 'health-iving',
 'health--beauty',
 'disposables',
 'kitche-hop',
 'tamim-roducts']
Name= []
Price=[]
Brand=[]
Offer=[]
scrol=0
for index in Category:

    for scrol in range(100): # for loop to get through all the pages by Assume the maximum number of pages for a category 
        try:
            scrol=scrol+1
            response = requests.get("https://shop.tamimimarkets.com/category/"+index+"&layoutType=GRID&sorting=NEW_FIRST&page="+str(scrol))
            soup = BeautifulSoup(response.text, 'html.parser')
            #print(response.status_code)
            x=0
            allProduct = soup.find(class_="ProductCollection__StyledContainer-sc-33fjso-2 jgRiGw")
            product = allProduct.find_all(class_="Product__StyledA-sc-13egllk-0 lnZJxA")
            for x in range (10000): # for loop to get through all the products in page 
                try:
                    subProduct= product[x] 
                    name = [n.get_text() for n in subProduct.find_all(class_='Text-sc-1bsd7ul-0 eKTmJK')]
                    price = [p.get_text() for p in subProduct.find_all(class_='Text-sc-1bsd7ul-0 loTFgW')]
                    brand = [b.get_text() for b in subProduct.find_all(class_='Text-sc-1bsd7ul-0 fOTkAa')]
                    offer = [o.get_text() for o in subProduct.find_all(class_='Price__StyledDel-sc-1svix5t-0 iUgmeI')]
                    x=x+1
                    categoryName.append(index)
                    Name.append(name)
                    Price.append(price)
                    Brand.append(brand)
                    Offer.append(offer)

                # If the products end in this category
                except IndexError:
                    break
       # If the number of pages ends for this category
        except AttributeError:
            break
            
Product1 = pd.DataFrame({
    "Category": categoryName,
    "Name": Name,
    "Price": Price,
    "Brand": Brand,
    "Offer": Offer,
})        
Product1   
```

```python
 Product1.to_csv('TamimiMarkets.csv')  
```
